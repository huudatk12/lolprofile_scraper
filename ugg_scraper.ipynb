```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
from time import sleep
import csv

print('-Finish importing package-')
```

```python
#Step1: Login the ugg

#Open Chorme and login the ugg site
driver = webdriver.Chrome()
url = 'https://u.gg/login'
driver.get(url)
sleep(3)
```

```python
#Import username and password
credential = open('login.txt')
line = credential.readlines()
username = line[0]
password = line[1]
print('-Finish importing the login credentials')

#Key username
email_field = driver.find_element('xpath','//*[@id="content"]/div/form/div[1]/div[2]/input')
email_field.send_keys(username)
print('-Finish keying in username')
sleep(2)

#Key password
email_field = driver.find_element('xpath','//*[@id="content"]/div/form/div[2]/div[3]/input')
email_field.send_keys(password)
print('-Finish keying in password')
sleep(3)

#Bottom login
email_field = driver.find_element('xpath','//*[@id="content"]/div/form/div[4]')
email_field.click()
print('-Finish logging in')
sleep(3)
```

```python
#Step 2: Search for the  frofile we want to crawl
#Locate the search bar element
search_field = driver.find_element('xpath','//*[@id="side-nav"]/div[1]/div[2]/a[3]')
search_field.click()
print('Finish searching !')
```

```python
# Task 3: Scrape the URLs of the profiles

# Task 3.1: Write a function to extract the URLs of one page
def GetURL():
    page_source = BeautifulSoup(driver.page_source)
    profiles = page_source.find_all('a', class_ = ['summoner','summoner-name'])
    all_profile_URL = []
    for profile in profiles:
        profile_ID = profile.get('href')
        profile_URL = "https://u.gg" + profile_ID
        if profile_URL not in all_profile_URL:
            all_profile_URL.append(profile_URL)
    return all_profile_URL

# Task 3.2: Navigate through many page, and extract the profile URLs of each page

input_page = int(input('How many pages you want to scrape: '))
URLs_all_page = []
for page in range(input_page):
    URLs_one_page = GetURL()
    sleep(2)
    driver.execute_script('window.scrollTo(0, document.body.scrollHeight);') #scroll to the end of the page
    sleep(3)
    next_button = driver.find_element(By.CLASS_NAME,'skip-to')
    driver.execute_script("arguments[0].click();", next_button)
    URLs_all_page = URLs_all_page + URLs_one_page
    sleep(2)

print('- Finish Task 3: Scrape the URLs')
```

```python
# Task 4: Scrape the data of ugg profile, and write the data to a .CSV file

with open('output.csv', 'w',encoding='utf-8', newline = '') as file_output:
    headers = ['In game', 'Rank', 'Rank point', 'Win rate', 'URL']
    writer = csv.DictWriter(file_output, delimiter=',', lineterminator='\n',fieldnames=headers)
    writer.writeheader()
    for ugg_URL in URLs_all_page:
        driver.get(ugg_URL)
        print('- Accessing profile: ', ugg_URL)

        page_source = BeautifulSoup(driver.page_source, "html.parser")

        in_game = page_source.find('div',class_= 'summoner-name').get_text().strip() #Remove unnecessary characters 
        print('--- In game is: ', in_game)
        rank_game = page_source.find('div',class_= 'rank-text')
        rank_name = rank_game.find('span').get_text()
        print('--- Rank is: ', rank_name)
        rank_point = rank_game.select_one(':nth-child(2)').get_text()
        print('--- Rank point is: ', rank_point)
        rate_game = page_source.find('div',class_= 'rank-wins')
        win_rate = rate_game.select_one(':nth-child(2)').get_text()
        print('--- Win rate is: ', win_rate)
        writer.writerow({headers[0]:in_game, headers[1]:rank_name, headers[2]:rank_point, headers[3]:win_rate, headers[4]:ugg_URL})
        print('\n')

print('Mission Completed!')
```
